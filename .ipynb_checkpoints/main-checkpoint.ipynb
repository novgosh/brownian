```python
def to_xyz(array):
    alpha, r, h = array
    return np.array([array[1] * np.cos(array[0]), array[1] * np.sin(array[0]), h])

def generate_population(size, min_d, max_d):
    if not size:
        return np.zeros((size, 4))
    
    ds = np.random.uniform(min_d, max_d, [size, 1])

    hs = np.random.uniform(ds * 0.5, H - ds * 0.5, [size, 1])
    alphas = np.random.uniform(0, np.pi * 2, [size, 1])
    rs = (R - ds * 0.5) * np.sqrt(np.random.uniform(0, 1, [size, 1]))
    
    coords = np.apply_along_axis(to_xyz, 1, np.hstack([alphas, rs, hs]))
    
    return np.hstack([coords, ds])

def generate_population_obsticles(population_size, obsticles_count):
    obsticles = generate_population(obsticles_count, O_MIN_D, O_MAX_D)
    population = []
    for i in range(population_size):
        being = None
        while being is None or np.sum(filter_collisions(being, obsticles)) > 0:
            being = generate_population(1, MIN_D, MAX_D)
        population.append(being)
        
    return np.vstack(population), obsticles

def generate_displacement(generation, delta_t):
    sigmas = np.sqrt(2 * preD / generation[:, 3] * delta_t)
    return np.random.randn(generation.shape[0], generation.shape[1] - 1) * sigmas[:, np.newaxis]

def stay_in_bounds(array):
    z_pos = array[2] + array[-1]
    sqr_r_pos = (array[0] + array[-3]) ** 2 + (array[1] + array[-2]) ** 2
    d2 = array[3] / 2
    
    if 0 <= z_pos - d2 and z_pos + d2 <= H and sqr_r_pos <= (R - d2) ** 2:
        array[:3] += array[-3:]
    return array[:4]

def check_for_out_of_bounds(being):
    first = population[:, N] <= population[:, N - 1]
    second = population[:, N - 1] <= H - population[:, N]
    third = np.linalg.norm(population[:, :N - 1], axis=1) <= R - population[:, N]
    
    return np.logical_not(np.logical_and(first, np.logical_and(second, third)))

def check_for_collisions(being, obsticles):
    return np.linalg.norm(obsticles[:, :N] - being[:N], axis=1) ** 2 < (obsticles[:, N] + being[N]) ** 2
    
def filter_collisions(population, obsticles):
    return np.any(np.apply_along_axis(check_for_collisions, 1, population, obsticles), axis=1)

def check_for_collisions_new(population, obsticle):
    return np.linalg.norm(population[:, :N] - obsticle[:N], axis=1) ** 2 < (population[:, N] + obsticle[N]) ** 2

def filter_collisions_new(population, obsticles):
    return np.any(np.apply_along_axis(check_for_collisions, 1, obsticles, population), axis=0)
    
def build_tracks(population, obsticles, length):
    tracks = np.zeros([population.shape[0], length, N])
    
    for i in range(length):
        tracks[:, i, :] = population[:, :N]
    
        deltas = generate_displacement(population, DELTA_T)
        population[:, :N] += deltas
#         print(filter_collisions(population, obsticles) == filter_collisions_new(population, obsticles))
#         assert np.all(filter_collisions(population, obsticles) == filter_collisions_new(population, obsticles))
        need_to_revert = np.logical_or(check_for_out_of_bounds(population), filter_collisions(population, obsticles))
#         need_to_revert = np.logical_or(check_for_out_of_bounds(population), filter_collisions_new(population, obsticles))

        population[:, :N][need_to_revert] -= deltas[need_to_revert]
        
    return tracks

def remove_same(track):
    indexes = np.hstack([[True], np.any(track[1:, :] != track[:-1, :], axis=1)]) 
    return track[indexes]

N = 3
ETA = 1.0e-3
kB = 1.38e-23
T = 293

preD = kB * T / (3 * np.pi * ETA)

R = 10 * 1e-6
H = 1.5 * 1e-5
MIN_D = 1e2 * 1e-9
MAX_D = MIN_D#3e2 * 1e-9
O_MIN_D = 4 * MAX_D * 5
O_MAX_D = 4 * MAX_D * 5

MAX_D_COEF = preD / MIN_D
DELTA_T = R**2 / (2 * (N - 1) * MAX_D_COEF) / 100000

PART_CNT = 1000
OBST_CNT = 300
TRACK_SIZE = 100

population, obsticles = generate_population_obsticles(PART_CNT, OBST_CNT)

start_time = time.time()
tracks = build_tracks(population, obsticles, TRACK_SIZE)
end_time = time.time()

print('{} to simulate {} partitions with {} obsticles'
      .format(end_time - start_time, PART_CNT, OBST_CNT))
print(np.mean(np.linalg.norm(tracks[:, 0, :] - tracks[:, -1, :], axis=1) ** 2))

fig = plt.figure(figsize=(20, 20))
ax = fig.gca()
ax.add_artist(plt.Circle((0, 0), R, color='b'))
for obsticle in obsticles:
    ax.add_artist(plt.Circle(obsticle[:N-1], obsticle[N]))
for (i, track) in enumerate(tracks):
    ax.plot(track[:, 0], track[:, 1])
#     if i > 100:
#         break
plt.show()
```
