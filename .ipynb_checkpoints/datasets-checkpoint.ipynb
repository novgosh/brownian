```python
def row_to_array(row):
    return list(map(lambda cell: float(cell.value) if cell.value else cell.value, row))

def is_header(row, template):
    for (value, cell) in zip(template, row):
        if value != cell.value:
            return False
    return True
```

```python
ARSNOC_HEADER = ['a', 'x', 'y'] 

def parse_arsnoc_sheet(sheet):
    track = []
    
    is_first = None
    
    for pos in range(sheet.nrows):
        row = sheet.row(pos)
        if is_first is None:
            if is_header(row, ARSNOC_HEADER):
                is_first = True
        else:
            row = row_to_array(row[:3])
            if is_first:
                r = row[0]
                is_first = False
            track.append([r] + row[1:])
    return np.array(track)        

def parse_arsnoc_book(book):
    result = []
    for sheet in book.sheets():
        current_track = parse_arsnoc_sheet(sheet)
        if current_track.shape[0] != 0:
            result.append(current_track)
            print(current_track.shape)
            
    return [result] 
        
data = []
for i in range(2, 15):
    file_name = './data/arsnoc{}cnfoc.xls'.format(i)
    print(file_name)
    book = xlrd.open_workbook(file_name)
    data.extend(parse_arsnoc_book(book))
```

```python
HEADER   = ['клетка 1', "Area", "XM", "YM", "XStart", "YStart"]
NEW_CELL = ['клетка 2', '', '', '', '', '']


def is_empty(row):
    for cell in row:
        if cell.value:
            return False
    return True

def is_new_cell(row):
    for i, cell in enumerate(row):
        if NEW_CELL[i] != cell.value:
            return False
    return True

def parse_sheet(sheet):
    data = []
    cell = []
    track = []
    
    assert is_header(sheet.row(0), HEADER)
    
    for pos in range(1, sheet.nrows):
        row = sheet.row(pos)
        if is_empty(row):
            if track:
                cell.append(np.array(track))
                track = []
        elif is_new_cell(row):
            if track:
                cell.append(np.array(track))
                track = []

            data.append(cell)
            cell = []
        else:
            track.append(row_to_array(row))
    if track:
        cell.append(np.array(track))
        track = []

    if cell:
        data.append(cell)
        
    return data

def parse_new_tracks(book):
    data = []
    for sheet in book.sheets():
        data.extend(parse_sheet(sheet))
    return data
```

```python
book = xlrd.open_workbook('./data/tracks_new.xls')
data = parse_new_tracks(book)

print('Data Length: {}'.format(len(data)))
for cell in data:
    print('\tCell Length: {}'.format(len(cell)))        
```

```python
for cell in data:
    length = np.amax(list(map(lambda track: track.shape[0], cell)))

    ys, ss = [], []
    for i in range(length // 2):
        cur = []
        for track in cell:
            if track.shape[0] > i:
                cur.append(np.sum((track[0, 2:4] - track[i, 2:4]) ** 2))
        ys.append(np.mean(cur))
        ss.append(len(cur))
    plot(arange(len(ys)), ys)
    figure()
```

```python
def t_function(x, alpha, beta):
    return beta * x**alpha

for cell in data:
    for track in cell:
        ys, ss = [], []
        for i in range(track.shape[0]):
            cur = []
            for j in range(track.shape[0] - i):
                cur.append(np.sum((track[j, 1:3] - track[j + i, 1:3]) ** 2))
            ys.append(np.mean(cur))
            ss.append(len(cur))
        ys = np.array(ys)
        
        xs = arange(len(ys))[:len(ys) // 2]
        ys = ys[xs]
            
        popt, _ = spo.curve_fit(t_function, xs, ys, [1, 1])
        
        print(popt)
        plot(xs, ys)
        plot(xs, t_function(xs, *popt))
        show()
```
